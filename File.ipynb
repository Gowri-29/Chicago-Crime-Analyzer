```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot(df, column):
    # Create subplots
    fig, axes = plt.subplots(1, 3, figsize=(20, 5))

    # Box plot
    sns.boxplot(data=df, x=column, ax=axes[0])
    axes[0].set_title(f'Box Plot for {column}')

    # Distribution plot
    sns.histplot(data=df, x=column, kde=True, bins=50, ax=axes[1])
    axes[1].set_title(f'Distribution Plot for {column}')

    # Violin plot
    sns.violinplot(data=df, x=column, ax=axes[2])
    axes[2].set_title(f'Violin Plot for {column}')

    # Display the figure
    plt.tight_layout()
    plt.show()
```

```python
df_cat = df.select_dtypes(object)
column_Cat_names = df_cat.columns.tolist()

df_encoded = df.copy()

def encode_cat(df, column_name):
    from sklearn.preprocessing import LabelEncoder

    # Instantiate the LabelEncoder object
    le = LabelEncoder()

    # Fit and transform the specified column
    encoded_column = le.fit_transform(df[column_name])

    # Replace the original column with the encoded values
    df[column_name] = encoded_column

    # Return the modified DataFrame
    return df

for column_name in column_Cat_names:
    df_encoded = encode_cat(df_encoded, column_name)
```

```python
import numpy as np
```

```python
#Skew removing:
df_skewed = df_encoded.copy()
df_skewed['Block'] = np.log(df_skewed['Block'])
df_skewed['IUCR'] = np.log(df_skewed['IUCR'])
df_skewed['Primary Type'] = np.log(df_skewed['Primary Type'])
df_skewed['Description'] = np.log(df_skewed['Description'])
df_skewed['Location Description'] = np.log(df_skewed['Location Description'])
df_skewed['Beat'] = np.log(df_skewed['Beat'])
df_skewed['District'] = np.log(df_skewed['District'])
df_skewed['Ward'] = np.log(df_skewed['Ward'])
df_skewed['Community Area'] = np.log(df_skewed['Community Area'])
df_skewed['FBI Code'] = np.log(df_skewed['FBI Code'])
df_skewed = df_skewed.dropna()
```

```python
df_skewed.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop rows with NaN values
df_skewed.dropna(inplace=True)
```

```python
outlier_column = ["IUCR","Primary Type","Description","Location Description","FBI Code"]
df_outiler = df_skewed.copy()
```

```python

def outlier(df, column):
    iqr = df[column].quantile(0.75) - df[column].quantile(0.25)
    upper_threshold = df[column].quantile(0.75) + (1.5 * iqr)
    lower_threshold = df[column].quantile(0.25) - (1.5 * iqr)
    df[column] = df[column].clip(lower_threshold, upper_threshold)
    return df

# Apply the function to numerical columns
for col in outlier_column:
    df_outlier = outlier(df_outiler, col)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, classification_report

def machine_crime(X, y, algorithm):   
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
    
    # Initialize the model with specified algorithm
    model = algorithm()
    
    # Fit the model on the training data
    model.fit(X_train, y_train)
    
    # Predict on the test set
    y_pred = model.predict(X_test)
    
    # Calculate accuracy for training and testing sets
    train_accuracy = accuracy_score(y_train, model.predict(X_train))
    test_accuracy = accuracy_score(y_test, y_pred)
    
    # Calculate F1 score
    f1 = f1_score(y_test, y_pred)
    
    # Generate classification report
    classification_report_str = classification_report(y_test, y_pred)
    
    # Print evaluation metrics
    print("Train Accuracy:", train_accuracy)
    print("Test Accuracy:", test_accuracy)
    print("F1 Score:", f1)
    print("Classification Report:\n", classification_report_str)
```

```python
def confusion(X, y, algorithm):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)

    # Initialize the model with specified algorithm
    model = algorithm()

    # Fit the model on the training data
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    # Compute confusion matrix
    conf_matrix = confusion_matrix(y_test, y_pred)

    # Plot confusion matrix
    plt.figure(figsize=(7, 5))
    sns.heatmap(conf_matrix, annot=True)
    plt.xlabel('Predicted')
    plt.ylabel('Truth')
    plt.title('Confusion Matrix')
    
    # Display the plot
    plt.show()
    
confusion(X,y,LogisticRegression)    
```
